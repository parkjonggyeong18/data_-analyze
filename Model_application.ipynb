```python
from ipywidgets import widgets
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
total_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
def preprocessing(df):

    conversion_dict = {'Ex': 9, 'Gd': 7,'TA': 5,'Fa': 3,'Po': 1,'NA': 0}
    df['BsmtCond_num'] = df['BsmtCond'].fillna(df['BsmtCond'].mode()[0]).map(conversion_dict)
    df['HeatingQC_num'] = df['HeatingQC'].fillna(df['HeatingQC'].mode()[0]).map(conversion_dict)
    df['KitchenQual_num'] = df['KitchenQual'].fillna(df['KitchenQual'].mode()[0]).map(conversion_dict)
    df['GarageQual_num'] = df['GarageQual'].fillna(df['GarageQual'].mode()[0]).map(conversion_dict)
    df['ExterQual_num'] = df['ExterQual'].fillna(df['ExterQual'].mode()[0]).map(conversion_dict)
    df['ExterCond_num'] = df['ExterCond'].fillna(df['ExterCond'].mode()[0]).map(conversion_dict)

    df['LotShape'] = df['LotShape'].replace({'Reg':4, 'IR1':3, 'IR2':2, 'IR3':1})
    df['Utilities'] = df['Utilities'].replace({'AllPub':4, 'NoSewr':3, 'NoSeWa':2, 'ELO':1})
    df['HouseStyle'] = df['HouseStyle'].replace({'1Story': '1Story', '1.5Fin': '1Story', '1.5Unf': '1Story',
                                                                 '2Story': '2Story', '2.5Fin': '2Story', '2.5Unf': '2Story',
                                                                 'SFoyer': 'Split', 'SLvl': 'Split'})

    one_hot_encoding = pd.get_dummies(df['HouseStyle'], prefix='HouseStyle')

    df = pd.concat([df, one_hot_encoding],axis=1)

    columns_to_drop = ['Id','MSSubClass', 'MSZoning', 'LotFrontage', 'Street',
                       'Alley', 'LandContour', 'LotConfig', 'LandSlope',
                       'Neighborhood', 'Condition1', 'Condition2', 'BldgType',
                       'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd',
                       'MasVnrType', 'MasVnrArea', 'Foundation', 'BsmtQual',
                       'BsmtExposure', 'BsmtFinType1', 'BsmtFinSF1', 'BsmtFinType2',
                       'BsmtFinSF2', 'BsmtUnfSF', 'Heating', 'CentralAir',
                       'Electrical', 'LowQualFinSF', 'BsmtHalfBath', 'HalfBath',
                       'Functional', 'FireplaceQu', 'GarageType', 'GarageYrBlt',
                       'GarageFinish', 'GarageCars', 'GarageCond', 'PavedDrive',
                       'PoolArea', 'PoolQC', 'Fence', 'MiscFeature', 'MiscVal',
                       'MoSold', 'YrSold', 'SaleType', 'SaleCondition', 'HouseStyle',
                       'ExterQual', 'ExterCond', 'BsmtCond', 'HeatingQC',
                       'KitchenQual', 'GarageQual']

    # y_train = total_data.iloc[:,-1]
    df.drop(columns=columns_to_drop,axis=1,inplace=True)
    return df
```

```python
total_data = preprocessing(total_data)
total_data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
from sklearn import linear_model
from sklearn.ensemble import GradientBoostingRegressor
```

```python
y = total_data['SalePrice']
X = total_data.drop(['SalePrice'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
model = GradientBoostingRegressor(n_estimators=250, learning_rate=0.08, random_state=10)
model.fit(X_train, y_train)
```

```python
X.mean()
```

```python
style = {'description_width': 'initial'}

inputLotArea = widgets.IntText(description='Lot Area: ', 
                               style=style, 
                               disabled=False)

inputYearBuilt = widgets.IntText(description='Built Year: ', 
                                 style=style, 
                                 disabled=False)

inputRemodAdd = widgets.IntText(description='Remodel Add Year: ', 
                                style=style, 
                                disabled=False)

inputBsmtSF = widgets.IntText(description='Basement Square Feet: ', 
                              style=style, 
                              disabled=False)

inputLivArea = widgets.IntText(description='Live Area Square Feet: ', 
                               style=style, 
                               disabled=False)

inputBathroom = widgets.IntText(description='Bathroom number: ', 
                                style=style, 
                                disabled=False)

inputBedroom = widgets.IntText(description='Bedroom number: ', 
                               style=style, 
                               disabled=False)

inputGarageArea = widgets.IntText(description='Garage Area: ', 
                                  style=style, 
                                  disabled=False)

inputFireplace = widgets.Dropdown(description='Include Fireplace: ',
                                  options=[('No', 0), ('Yes', 1)],
                                  value=0,
                                  style=style,
                                  disabled=False)

inputHouseStyle = widgets.Dropdown(description='House Style: ',
                                   options=[('One story/One and one-half story', 0), ('Two story/Two and one-half story', 1), ('Split Foyer', 2)],
                                   value=0,
                                   style=style,
                                   disabled=False)

out1 = widgets.Output(layout={'border': '1px solid black'})

def onclick_input(tmp):
    out1.clear_output()
    input_df = pd.DataFrame(columns=X.columns)
    input_df.loc[0] = X.mean()
    input_df.loc[0, 'LotArea'] = inputLotArea.value
    input_df.loc[0, 'YearBuilt'] = inputYearBuilt.value
    input_df.loc[0, 'YearRemodAdd'] = inputRemodAdd.value
    input_df.loc[0, 'TotalBsmtSF'] = inputBsmtSF.value
    input_df.loc[0, 'GrLivArea'] = inputLivArea.value
    input_df.loc[0, 'FullBath'] = inputBathroom.value
    input_df.loc[0, 'BedroomAbvGr'] = inputBedroom.value
    input_df.loc[0, 'GarageArea'] = inputGarageArea.value
    input_df.loc[0, 'Fireplaces'] = inputFireplace.value
    houseStyle = inputHouseStyle.value
    if houseStyle == 0:
        input_df.loc[0, 'HouseStyle_1Story'] = 1
        input_df.loc[0, 'HouseStyle_2Story'] = 0
        input_df.loc[0, 'HouseStyle_Split'] = 0
    elif houseStyle == 1:
        input_df.loc[0, 'HouseStyle_1Story'] = 0
        input_df.loc[0, 'HouseStyle_2Story'] = 1
        input_df.loc[0, 'HouseStyle_Split'] = 0
    elif houseStyle == 2:
        input_df.loc[0, 'HouseStyle_1Story'] = 0
        input_df.loc[0, 'HouseStyle_2Story'] = 0
        input_df.loc[0, 'HouseStyle_Split'] = 1
    pred = model.predict(input_df)
    with out1:
        print(f'{pred[0]:.2f}USD')
        
box_input1=widgets.Box([inputLotArea, inputYearBuilt])
box_input2=widgets.Box([inputRemodAdd, inputBsmtSF])
box_input3=widgets.Box([inputLivArea, inputBathroom])
box_input4=widgets.Box([inputBedroom, inputGarageArea])
box_input5=widgets.Box([inputFireplace, inputHouseStyle])

button = widgets.Button(description='Submit',
                        disabled=False)
button.on_click(onclick_input)
display(box_input1, box_input2, box_input3, box_input4, box_input5, button, out1)
```
